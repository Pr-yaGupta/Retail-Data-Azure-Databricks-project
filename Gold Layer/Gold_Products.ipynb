### **DLT Pipeline**

```python
import dlt
from pyspark.sql.functions import *
```

**Streaming Table**

```python
# Expectations
rules = {
    "rule1" : "product_id IS NOT NULL",
    "rule2" : "product_name IS NOT NULL"
}
```

```python
# Decorator
@dlt.table()

# Creating materialized view on top of the table
@dlt.expect_all_or_drop(rules)
def dim_products_stage():
  df = spark.read.table("databricks_catalog.silver.products")
  return df
```

**Streaming View**

```python
@dlt.view()

def dim_products_view():
    df = spark.readStream.table("Live.dim_products_stage")
    return df
```

**Dim_Products**

```python
dlt.create_streaming_table("dim_products")

dlt.apply_changes(
  target = "dim_products",
  source = "Live.dim_products_view",
  keys = ["product_id"],
  sequence_by = "product_id",
  stored_as_scd_type = 2
)
```
